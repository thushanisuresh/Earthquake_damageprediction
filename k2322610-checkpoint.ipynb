<b>INTRODUCTION<b> 

Natural catastrophes, especially earthquakes, put infrastructure at serious risk, putting lives in jeopardy and resulting in large financial damages. A multitude of variables, such as the age, structural design, and location of the structure, as well as the materials used during construction, may significantly affect how severely an earthquake damages a building. Reducing possible damages in earthquake-prone locations, raising safety standards, and strengthening building resilience all depend on an understanding of these elements.

The main emphasis of this paper is a thorough exploratory data analysis (EDA) of an extensive dataset that records different building attributes and the degree of seismic damage they sustained. The dataset offers a singular chance to investigate the link between building traits and earthquake susceptibility. It includes structural elements, materials employed, and other pertinent building metrics. We want to forecast the potential extent of a building's damage in the case of an earthquake by using this data, which is crucial for risk assessment and mitigation plans.

The primary objectives of this analysis are to:
1.	Identify key factors that significantly influence the earthquake damage grade of buildings.
2.	Develop a predictive model to estimate potential damage outcomes based on building characteristics.
3.	Provide actionable insights and recommendations to stakeholders, including construction companies, policy makers, and disaster response teams, to inform better building practices and disaster preparedness plans.


<b>INTRODUCTION TO DATASET</b>

Dataset Overview
This dataset provides detailed information about buildings and their damage levels following an earthquake. Comprised of two main files, input_features.csv and target_values.csv, the dataset captures a wide range of structural and locational attributes for each building.

Features:

Structural Characteristics: Includes data on the number of floors, age, area, and height of the buildings.
Construction Materials: Details the materials used in the superstructure such as timber, bamboo, adobe mud, and various types of mortar.
Location: Described by three geographic levels from broad regions to specific sub-regions.
Additional Uses: Information on secondary uses of the building like agricultural, hotel, or institutional purposes.
Target Variable:

Damage Grade: An ordinal variable that categorizes the extent of damage sustained by the buildings into three grades: low, medium, and high damage.

<h4> Function to Calculate File Size üìã</h4>

```python
def get_file_size(file_path):
    """Return the file size in bytes and megabytes."""
    file_size_bytes = os.path.getsize(file_path)
    file_size_mb = file_size_bytes / (1024 * 1024)
    return file_size_mb, file_size_bytes
```

<h4>Function to Calculate SHA1 Checksum</h4>

```python
def get_checksum(file_path):
    """Calculate the SHA1 checksum of a file."""
    sha1 = hashlib.sha1()
    with open(file_path, "rb") as f:
        while True:
            data = f.read(65536)  # Read the file in chunks of 65536 bytes
            if not data:
                break
            sha1.update(data)
    return sha1.hexdigest()
```

```python
def get_file_extension(file_path):
    """Extract and return the file extension."""
    return os.path.splitext(file_path)[1][1:]
```

<h4>Analyze File Properties</h4>

```python
def display_file_properties(file_path):
    file_name = os.path.basename(file_path)
    file_size_mb, file_size_bytes = get_file_size(file_path)
    file_extension = get_file_extension(file_path)
    checksum = get_checksum(file_path)
    
    print(f"File Name: {file_name}")
    print(f"File Size: {file_size_mb:.2f} MB ({file_size_bytes} bytes)")
    print(f"File Type: {file_extension.upper()}")
    print(f"SHA1 Checksum: {checksum}")
    print()  # Blank line for better readability between files
```

<h4>Building Use and Vulnerability to Earthquake Damage</h4>
